## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
def generate_random(bottom:int, ceiling:int, count:int)->list:
    """
    Generate uniform random numbers between the given bounds.

    Parameters:
    bottom (float): The lower bound of the range.
    ceiling (float): The upper bound of the range.
    count (int): The number of random numbers to generate.

    Returns:
    numpy.ndarray: An array of uniformly distributed random numbers.
    """
    return scipy.stats.uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count)
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

```python
from scipy.stats import norm 
```

```python
def generate_normal_random(mean:int, std_dev:float, size:int)->list:
    """
    Generate normally distributed random numbers.

    Parameters:
    mean (float): Mean (average) of the distribution.
    std_dev (float): Standard deviation (spread or dispersion) of the distribution.
    size (int): Number of random numbers to generate.

    Returns:
    numpy.ndarray: An array of normally distributed random numbers.
    """
    return stats.norm.rvs(loc=mean, scale=std_dev, size=size)
```

```python
norm1=generate_normal_random(10,1,1000)
norm2=generate_normal_random(10,50,1000)
```

```python
#Set of numbers norm1
mean=10
std_dev=1

# Plotting the histogram of the generated random numbers
plt.figure(figsize=(8, 6))
plt.hist(norm1, bins=30, density=True, edgecolor='black')

# Plotting the theoretical normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = scipy.stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Distribution 1')
plt.legend(['Normal Distribution', 'Generated Data'])
plt.grid(True)
plt.show()
```

```python
#Set of numbers norm1
mean=10
std_dev=50

# Plotting the histogram of the generated random numbers
plt.figure(figsize=(8, 6))
plt.hist(norm2, bins=30, density=True, edgecolor='black')

# Plotting the theoretical normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = scipy.stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Distribution 2')
plt.legend(['Normal Distribution', 'Generated Data'])
plt.grid(True)
plt.show()
```

```python
#KS test for Combined 
from scipy.stats import kstest, norm
import numpy as np
data = df['Combined MPG']

#Perform KS test against a normal distribution
statistic, p_value = kstest(data, 'norm')

# Print the results
print(f'KS Statistic: {statistic}')
print(f'P-value: {p_value}')

alpha = 0.05
if p_value > alpha:
    print('Sample distribution is likely normal (fail to reject H0)')
else:
    print('Sample distribution is not normal (reject H0)')
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

```python
import numpy as np

def generate_exponential_numbers(mean, size):
    # Calculate the scale parameter for numpy's exponential distribution
    scale = 1 / mean
    
    # Generate exponentially distributed numbers with the specified scale and size
    exponential_numbers = np.random.exponential(scale=scale, size=size)
    
    return exponential_numbers


exponential_data_10= generate_exponential_numbers(10, 10)
exponential_data_100= generate_exponential_numbers(10, 100)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(exponential_data_10, bins=100, edgecolor='black')
plt.xlabel('data 10')
plt.ylabel('Frequency')
plt.title('Histogram of data 10')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
plt.hist(exponential_data_100, bins=100, edgecolor='black')
plt.xlabel('data 100')
plt.ylabel('Frequency')
plt.title('Histogram of data 100')
plt.grid(True)
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well.
