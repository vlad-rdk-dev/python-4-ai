Please note that this time the order of the fruits being picked does not matter.

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
from scipy.stats import binom

# Given data
total_fruits = 100
apples = 60
sample_size = 20

# Probability of picking an apple
prob_apple = apples / total_fruits

# Calculate cumulative probability P(X < 5)
probability_less_than_5_apples = binom.cdf(4, sample_size, prob_apple)

print(f"Probability that less than 5 fruits picked are apples: {probability_less_than_5_apples:.6f}")
````

**Plot the PDF (probability distributiom function) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Given data
total_fruits = 100
apples = 60
sample_size = 20

# Probability of picking an apple
prob_apple = apples / total_fruits

# Generate x values (number of apples)
x = np.arange(0, sample_size + 1)  # From 0 to sample_size inclusive

# Calculate PDF values for the binomial distribution
pdf_values = binom.pmf(x, sample_size, prob_apple)

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.bar(x, pdf_values, align='center', alpha=0.7, color='blue')
plt.xticks(x)
plt.xlabel('Number of Apples')
plt.ylabel('Probability')
plt.title('Binomial Distribution PDF: Number of Apples in 20 Fruit Sample')
plt.grid(True)
plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
import math

# Given data
average_goals = 2.3
k = 5

# Calculate lambda^k * e^(-lambda)
poisson_part = (average_goals ** k) * math.exp(-average_goals)

# Calculate k!
factorial_k = math.factorial(k)

# Calculate P(X = 5)
probability_5_goals = poisson_part / factorial_k

print(f"The probability that exactly 5 goals are scored in a match is: {probability_5_goals:.6f}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
import math
import matplotlib.pyplot as plt

# Given data
average_goals = 2.3
k_values = range(0, 11)  # Values of k from 0 to 10

# Calculate Poisson probabilities for each k
poisson_probs = []
for k in k_values:
    # Calculate lambda^k * e^(-lambda) / k!
    poisson_prob = (average_goals ** k) * math.exp(-average_goals) / math.factorial(k)
    poisson_probs.append(poisson_prob)

# Plot the Poisson distribution
plt.figure(figsize=(10, 6))
plt.bar(k_values, poisson_probs, color='blue', alpha=0.7)
plt.title('Poisson Probability Distribution: Number of Goals')
plt.xlabel('Number of Goals')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(True)
plt.show()
````
